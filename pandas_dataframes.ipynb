```python
import numpy as np
import pandas as pd
```

### Creating DataFrame

```python
# using lists
student_data = [
    ['sikandar',100,80,10],
    ['umair',90,70,7],
    ['hassan',120,100,14],
    ['ali',80,50,2]
]

pd.DataFrame(student_data,columns=['name','iq','marks','package'])
```

```python
# using dicts

student_dict = {
    'name':['nitish','ankit','rupesh','rishabh','amit','ankita'],
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}

students = pd.DataFrame(student_dict)
students.set_index('name',inplace = True)
students
```
